# SQL Queries

```python
#List all tables in the database
# Importing Sqlite3 Module
def execute_query(query=''):
    
    import sqlite3

    try:

        # Making a connection between sqlite3 
        # database and Python Program
        sqliteConnection = sqlite3.connect('paintings.db')

        # If sqlite3 makes a connection with python
        # program then it will print "Connected to SQLite"
        # Otherwise it will show errors
        print("Connected to SQLite")

        # Getting all tables from sqlite_master
      

        # Creating cursor object using connection object
        cursor = sqliteConnection.cursor()

        # executing our sql query
        cursor.execute(query)
        print("List of tables\n")
       
        # printing all tables list
        col_name=[i[0] for i in cursor.description]
        
        data = cursor.fetchall()
        df = pd.DataFrame(data, columns=col_name)
        print(df)

    except sqlite3.Error as error:
        print("Failed to execute the above query", error)

    # finally:

    # 	# Inside Finally Block, If connection is
    # 	# open, we need to close it
    # 	if sqliteConnection:

    # 		# using close() method, we will close 
    # 		# the connection
    # 		sqliteConnection.close()

    # 		# After closing connection object, we 
    # 		# will print "the sqlite connection is 
    # 		# closed"
    # 		print("the sqlite connection is closed")

```

```python
#1. Fetch all the paintings which are not displayed at any museums
res = execute_query(query="""SELECT *
FROM work
WHERE museum_id IS NULL
LIMIT 10;""")
```

```python
#2. Are there any museums without any paintings?
res = execute_query(query="""SELECT *
FROM museum AS m
LEFT JOIN work AS w
    ON m.museum_id = w.museum_id
    WHERE w.museum_id IS NULL;""")
```

```python
#3.How many paintings have an asking price which more than their regular price?
res = execute_query(query="""SELECT
    COUNT(*) AS no_of_paintings
FROM product_size
WHERE sale_price > regular_price;""")
```

```python
#4. Identify the paintings whose asking price is less than 50% of its regular price
res = execute_query(query="""SELECT *
FROM product_size
WHERE sale_price < (0.50 * regular_price)
LIMIT 20;""")
```

```python
#5. Which canvas size costs the most?
res = execute_query(query="""SELECT DISTINCT(size_id),
  width,
  height,
  label
FROM (
  SELECT
    c.*,
    p.regular_price
  FROM canvas_size AS c
    INNER JOIN product_size AS p
    ON c.size_id = p.size_id
  WHERE regular_price = (SELECT MAX(regular_price) FROM product_size)) AS max_price;""")
```

```python
#7. Identify the museums which are open on both Sunday and Monday.
res = execute_query(query="""SELECT
  mh.museum_id,
  m.name,
  m.city
FROM museum_hrs AS mh
JOIN museum AS m 
  ON mh.museum_id = m.museum_id
WHERE day = 'Sunday' 
  AND mh.museum_id IN (SELECT museum_id FROM museum_hrs WHERE day = 'Monday');
""")
```

```python
#8. How many museums are open every single day?
res = execute_query(query="""SELECT 
  COUNT(*) AS everyday_museums
FROM(
  SELECT *,
    ROW_NUMBER() OVER(PARTITION BY museum_id) AS rn
  FROM museum_hrs) AS museum_hrs_days
  WHERE rn = 7;
""")
```

```python
#9. Which are the top 5 most popular museums?
res = execute_query(query="""SELECT  *
FROM(
  SELECT 
    m.name AS museum_name,
    COUNT(w.work_id) AS nr_of_paintings
  FROM museum AS m
  JOIN work AS w
    ON m.museum_id = w.museum_id
  GROUP BY 1) AS museum_painting_counts
ORDER BY nr_of_paintings DESC
LIMIT 5;
""")
```

```python
#10. Who are the top 5 most popular artists?
res = execute_query(query="""SELECT *
FROM (
  SELECT 
    a.full_name AS artist_name,
    COUNT(w.work_id) AS no_of_paintings
  FROM artist AS a
  JOIN work AS w
    ON a.artist_id = w.artist_id
  GROUP BY 1)AS artist_painting_counts
ORDER BY no_of_paintings DESC
LIMIT 5;
""")
```

```python
#11. Display the 3 least popular canvas sizes
res = execute_query(query="""SELECT *
FROM (
  SELECT
    c.label as canvas_label,
    COUNT(w.work_id) AS no_of_paintings
  FROM canvas_size AS c
  JOIN product_size AS p
    ON c.size_id = p.size_id
  JOIN work AS w
    ON p.work_id = w.work_id
  GROUP BY 1
)AS canvas_label_painting_counts
ORDER BY no_of_paintings ASC
LIMIT 3;
""")
```

```python
#12. Which museum is open for the longest during a day?
res = execute_query(query="""SELECT 
    m.name,
    m.city,
    mh.day,
    (mh.open - mh.close)  AS hrs_open
FROM museum_hrs AS mh
JOIN museum AS m
    ON mh.museum_id = m.museum_id
ORDER BY (mh.open - mh.close) DESC
LIMIT 1;""")
```

```python
#13. Identify the top 5 artists whose paintings are displayed in multiple countries
res = execute_query(query="""SELECT 
  a.full_name AS artist_name,
  COUNT(*) AS no_of_countries_displayed
FROM museum AS m
JOIN work AS w
  ON m.museum_id = w.museum_id
JOIN artist AS a
  ON w.artist_id = a.artist_id
GROUP BY 1
HAVING no_of_countries_displayed > 2
ORDER BY COUNT(*) DESC
LIMIT 5;""")
```
